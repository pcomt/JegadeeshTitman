```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

df = pd.read_csv("CRSP-data.csv")
```

Première étape : nettoyage des données 

On élimine les données manquantes en considérant que le dataset a assez de données pour enlever celles incomplètes. 
On élimine les actions cotées en dehors du NYSE et de l'AMEX.

```python
#Drop des NaN

df.dropna(axis=0,inplace=True)

#Drop des actions hors NYSE & AMEX

df = df[df["PRIMEXCH"].isin(["A","N"])]
```
